```python
from tensorflow import keras
```

```python
ac = keras.metrics.Accuracy()
ac.update_state([1,2,3,4],[1,0,3,4])
ac.result().numpy()
```

```python
mae = keras.metrics.MAE([10,12,18,13],[12,14,20,15])
mae.numpy()
```

```python
ba = keras.metrics.BinaryAccuracy()
ba.update_state([0,1,0,1,1,0],[0.1,0.9,0.4,.092,0.38,0.002])
ba.result().numpy()
```

```python
catogorical= keras.metrics.CategoricalAccuracy()
catogorical.update_state([[0,0,1,0,1],[1,0,0,1,0]],
                 [[0.1,0,1,0.12,0.81],[1,0.1,0,0.8,0]])
 
catogorical.result().numpy()
```

```python
cc = keras.metrics.CategoricalCrossentropy()
```

```python
cc.update_state([[0,1],[1,1]],[[0.6,0.5],[0.5,0.6]])
cc.result().numpy()
```
